# Afeter Analysing the data we will do the encoding process on to the categorical data as our model only understands the numerical values

```python
df_cat=df.iloc[:,:-1]
df_cat.drop(['BMI','PhysicalHealth','MentalHealth','SleepTime'],axis=1,inplace=True)
```

```python
df_cat
```

```python
df_num=df[['BMI','PhysicalHealth','MentalHealth','SleepTime']]
df_num
```

```python
from sklearn.preprocessing import LabelEncoder
for i in df_cat:
    le=LabelEncoder()
    df_cat[i]=le.fit_transform(df_cat[i])
```

```python
df_cat
```

```python
# after finishing the encodind process concat the categoricala nd numerical datset
```

```python
df2=pd.concat([df_cat,df_num],axis=1)
df2.head()
```

# After finishiing the encoding process we will split the data and and apply the train test split method

```python
x=df2.iloc[:,1:]
x
```

```python
y=df2.iloc[:,0]
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=1)
#to check shape of the dataset after spliting the dataset
xtrain.shape,xtest.shape,ytrain.shape,ytest.shape
```

# Now we will build the model for the dataset and we will work on two models 1. LogisticRegression 2.KNN

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import r2_score
from sklearn.metrics import classification_report
```

```python
#Using LogisticRegression Model
lr=LogisticRegression()
lr.fit(xtrain,ytrain)
ypred1=lr.predict(xtest)
acc1=accuracy_score(ytest,ypred1)
clsfc1=classification_report(ytest,ypred1)
print("Accuracy of model is:",acc1)
print("Classification report:",clsfc1)
```

```python
#Using KNN model
knn=KNeighborsClassifier()
knn.fit(xtrain,ytrain)
ypred2=knn.predict(xtest)
acc2=accuracy_score(ytest,ypred2)
clsfc2=classification_report(ytest,ypred2)
print("Accuracy of model is:",acc2)
print("Classification report:",clsfc2)
```
